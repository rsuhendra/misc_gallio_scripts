```python
df = pd.read_csv('temp.txt', delimiter = '\t', names=['chr', 'start', 'end', 'type', 'gie'])

k = 1000000 # centromere size

centro = {'chr': ['chr2R', 'chr2R'], 'start': [0, k], 'end': [k,2*k],  
          'type':['centro', 'centro'], 'gie':['acen', 'acen']}
df = pd.concat([df,pd.DataFrame(centro)], ignore_index = True)
df.reset_index()

df.loc[(df['type'] =='centro') & (df['chr'] == 'chr2R') , ['end','start']] += 22570260
df.loc[(df['type'] != 'centro') & (df['chr'] == 'chr2R') , ['end','start']] += 22570260 - 348419 + 2*k
df.loc[df['chr'] == 'chr2R' , 'chr'] = 'chr2'
df.loc[df['chr'] == 'chr2L' , 'chr'] = 'chr2'

centro = {'chr': ['chr3R', 'chr3R'], 'start': [0, k], 'end': [k,2*k], 
          'type':['centro', 'centro'], 'gie':['acen', 'acen']}
df = pd.concat([df,pd.DataFrame(centro)], ignore_index = True)
df.reset_index()

df.loc[(df['type'] =='centro') & (df['chr'] == 'chr3R') , ['end','start']] += 23849507
df.loc[(df['type'] != 'centro') & (df['chr'] == 'chr3R') , ['end','start']] += 23849507 - 0 + 2*k
df.loc[df['chr'] == 'chr3R' , 'chr'] = 'chr3'
df.loc[df['chr'] == 'chr3L' , 'chr'] = 'chr3'

df.loc[df['chr'] == 'chrX' , 'chr'] = 'chr1'

df = df.sort_values(by = ['chr', 'start'])

df.to_csv('test.txt', header= False, index = False, sep = '\t')
```
